```python
# Inheritance in Python

class Animal:
    def speak (self):
        return "Animal speak"

class Dog(Animal):
    def speak(self):
        return "Dog Barks"
    
class Fish(Animal):
    pass

animal = Dog()
print(animal.speak())

fish = Fish()
print(fish.speak())
```

```python
# Scenario 1: Create a User class with attributes and a method to display user details.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class User(Person):
    def __init__(self, name, age, email, city):
        super().__init__(name, age)   
        self.email = email
        self.city = city

    def display_user(self):
        print("User Details")
        print("Name :", self.name)
        print("Age :", self.age)
        print("Email :", self.email)
        print("City :", self.city)

u1 = User("Ayush", 21, "ayush@gmail.com", "Bhopal")
u1.display_user()
```

```python
# Scenario 2: Create a Product class and calculate discounted price using a method.

class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Product(Item):
    def __init__(self, name, price, discount_percent):
        super().__init__(name, price)
        self.discount_percent = discount_percent

    def discounted_price(self):
        discount_amount = (self.price * self.discount_percent) / 100
        final_price = self.price - discount_amount
        return final_price

    def display_product(self):
        print("Product Name :", self.name)
        print("Original Price :", self.price)
        print("Discount :", self.discount_percent, "%")
        print("Final Price :", self.discounted_price())

p1 = Product("Laptop", 50000, 10)
p1.display_product()
```

```python
# Scenario 3: Create a base class User and child class Admin that adds extra permissions.

class User:
    def __init__(self, username, email):
        self.username = username
        self.email = email

    def show_user(self):
        print("Username:", self.username)
        print("Email:", self.email)

class Admin(User):
    def __init__(self, username, email, permission):
        super().__init__(username, email)
        self.permission = permission

    def show_admin(self):
        print("\nAdmin Details")
        self.show_user()
        print("Permission:", self.permission)

a1 = Admin("AyushAdmin", "admin@gmail.com", "Full Access")
a1.show_admin()
```

```python
# Scenario 4: Base class Employee extended by Manager with extra responsibilities.

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_employee(self):
        print("Employee Name:", self.name)
        print("Salary:", self.salary)

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def show_manager(self):
        print("\nManager Details")
        self.show_employee()
        print("Department:", self.department)

m1 = Manager("Rohan", 60000, "IT Department")
m1.show_manager()
```
